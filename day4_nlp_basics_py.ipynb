```python
from nltk.tokenize import sent_tokenize
text = "Artificial Intelligence is transforming industries. NLP is a key part of AI. Tokenization is the first NLP step."
sent_tokenize(text)
```

```python
from nltk.tokenize import word_tokenize
text = "I am Avidi Sai kiran. Currently I am Pursuing Btech 2nd year in vishnu Institute of Technology."
word_tokenize(text)
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
word = ["playing", "played", "plays", "happily", "studies"]
stemmed_words = [stemmer.stem(w) for w in word]
print(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
word = ["running", "cars", "better", "children", "feet"]
lemmatized_words = [lemmatizer.lemmatize(w, "v") for w in word]
print(lemmatized_words)
```
